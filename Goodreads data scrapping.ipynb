```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
base_url="https://www.goodreads.com"
genres_list=[]
```

# List of Genres on website

```python
for i in range(1,15):
    # fetching reviews from five pages
    query_parameter = "/genres/list?page="+str(i)    
    url = base_url + query_parameter
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    #genres = soup.find_all('a', class_='gr-hyperlink') 
    
    genre_element = soup.findAll("div",attrs={"class","leftContainer"})
    #print(genre_element)
    
    for div_element in genre_element:
        genres=div_element.find_all('a', class_='mediumText actionLinkLite')
    
    for a_tag in genres:
        genres_list.append(a_tag.text)

    
print(genres_list)
```

```python
print(len(genres_list))
```

# Most common genres

```python
common_genres_list=[]
response = requests.get(base_url)
soup = BeautifulSoup(response.text, 'html.parser')

booklist_element = soup.findAll("div", id="browseBox")

for div_element in booklist_element:
    common_genres = div_element.find_all('a', class_='gr-hyperlink')
    
for a_tag in common_genres:
        common_genres_list.append(a_tag.text)
```

```python
print(common_genres_list)
```

# List of books from given list (user input) of genres

```python
ListOfGenres_UserInput=[str(i) for i in input("enter genres: ").split(',')]
```

```python
print(ListOfGenres_UserInput)
```

```python
number_of_books=int(input("enter number of books you need from each genre(max 50): "))
```

```python
all_books = []
all_books_organised = {}

for user_genre in ListOfGenres_UserInput:
    query_parameter = "/shelf/show/" + user_genre
    url = base_url + query_parameter
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    booklist_element = soup.findAll("div", attrs={"class", "leftContainer"})

    # Initialize the list for the current user_genre if it doesn't exist
    if user_genre not in all_books_organised:
        all_books_organised[user_genre] = []

    for div_element in booklist_element:
        fifty_books = div_element.find_all('a', class_='bookTitle')

    for i in range(number_of_books):
        book_name=re.sub(r'\([^)]*\)', '', fifty_books[i].text).strip()
        all_books.append(book_name)
        all_books_organised[user_genre].append(book_name)

# Now, all_books_organised contains lists of books organized by user_genre
```

```python
print(all_books_organised)
```

suggest a random book from your list?

```python
import random
random_book = random.choice(all_books)
print("Random Book:", random_book)
```
